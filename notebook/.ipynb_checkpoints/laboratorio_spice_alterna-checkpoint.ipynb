 Ahora extraemos los datos del archivo `.raw`.

```python
l = ltspice.Ltspice("files\corriente_alterna.raw")
l.parse()

tiempo = l.get_time()
vac = l.get_data('V(1)')
i_R1 = l.get_data('I(R1)')
i_VAC = l.get_data('I(V1)')
```

 Dibujamos las gráficas con los datos obtenidos con LTspice.

```python
plot.rcParams['figure.figsize'] = [6.4*1.7, 4.8]
plot.rcParams['font.size'] = 12

fig, ax = plot.subplots()
plot.title('Circuito Corriente Alterna - ($\mathrm{10v}$) - VAC:$\mathrm{120v}$-$\mathrm{60Hz}$ R:$\mathrm{10k\Omega}$')
ax.set(xlabel='Tiempo (s)', ylabel='Voltaje (V)')

line_iR1, = ax.plot(tiempo, i_R1)
line_iR1.set_label('Corriente R1')

line_iC2, = ax.plot(tiempo, i_VAC)
line_iC2.set_label('Corriente V1')

line_iR1, = ax.plot(tiempo, vac)
line_iR1.set_label('Voltaje VAC')

plot.legend()
plot.grid(True)
plot.tight_layout()
plot.show()
```
